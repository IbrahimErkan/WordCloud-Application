```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
oku = open("metin.txt").read()
oku
```

```python
wordcloud = WordCloud(background_color = "black",
                     width = 1200,
                     height = 1000).generate(oku)
plt.figure(figsize = (18,10))
plt.imshow(wordcloud)
plt.show()
```

```python
wordcloud = WordCloud(background_color = "white",
                     width = 1200,
                     height = 1000).generate(oku)
plt.figure(figsize = (20,10))
plt.imshow(wordcloud)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib
import random
```

```python
cmap = matplotlib.colors.LinearSegmentedColormap.from_list("",["#333333","#eeeeee"])

wordcloud = WordCloud(background_color = "white", width=800, height=400,
                     colormap = cmap, max_words = 150).generate(oku)
plt.figure(figsize = (18,10))
plt.imshow(wordcloud)
plt.axis("off")        #cerceve
plt.show()
```
